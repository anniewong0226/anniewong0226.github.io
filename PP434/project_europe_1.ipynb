```python
import pandas as pd

df = pd.read_csv('/Users/annie_lok_yan_wong/Downloads/europe_yearly_full_release_long_format.csv')
print(df.columns.tolist())
```

```python
countries_to_keep = [
    'NLD', 'DEU', 'FRA', 'ESP', 'ITA', 'GBR', 'BEL', 'POL', 'AUT', 
    'SWE', 'NOR', 'DNK', 'FIN', 'PRT', 'GRC', 'CZE', 'ROU', 'IRL', 'HUN'
]
filtered_df = df[df['ISO 3 code'].isin(countries_to_keep)]
filtered_df.head() 
```

```python
renewable_df = df[df['Variable'] == 'Clean']
```

```python
result = renewable_df.groupby('Year')['Value'].mean().reset_index()
result.columns = ['Year', 'Average Renewable Generation (%)']
print(result)
```

```python
result.to_csv('/Users/annie_lok_yan_wong/PP434/average_renewable_generation_europe.csv', index=False)
```

```python
clean_energy = df[(df['Subcategory'] == 'Aggregate fuel') & 
                  (df['Variable'] == 'Clean') & 
                  (df['Unit'] == '%')]

clean_by_country_year = clean_energy[['Area', 'ISO 3 code', 'Year', 'Value']]

clean_by_country_year.columns = ['Country', 'ISO Code', 'Year', 'Clean Energy (%)']

print(clean_by_country_year)

clean_by_country_year.to_csv('clean_energy_by_country_year.csv', index=False)
```
