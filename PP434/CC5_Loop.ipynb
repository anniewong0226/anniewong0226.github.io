```python
import requests
import json
import time
```

```python
API_KEY = '50PYVHBMHNN8UD6O'
BASE_URL = 'https://www.alphavantage.co/query'

symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA', 'NVDA']

for symbol in symbols:
    print(f"Downloading {symbol}...")

    params = {
        'function': 'TIME_SERIES_DAILY',
        'symbol': symbol,
        'apikey': API_KEY,
        'outputsize': 'compact'  
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    filename = f'{symbol}_data.json'
    with open(filename, 'w') as f:
        json.dump(data, f, indent=2)

    print(f"Saved {filename}")

    if symbol != symbols[-1]:  
        print("Waiting 12 seconds...")
        time.sleep(12)

print("All downloads complete!")
```

```python
import pandas as pd
```

```python

base_url = 'https://raw.githubusercontent.com/anniewong0226/anniewong0226.github.io/refs/heads/main/'
symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA', 'NVDA']

for symbol in symbols:
    print(f"Cleaning {symbol}...")
    
    url = f'{base_url}{symbol}_data.json'
    response = requests.get(url)
    data = response.json()
    
    time_series = data['Time Series (Daily)']
    df = pd.DataFrame.from_dict(time_series, orient='index')
    df.reset_index(inplace=True)
    df.rename(columns={'index': 'date'}, inplace=True)
    
    df_clean = df[['date', '1. open']].copy()
    df_clean.rename(columns={'1. open': 'open_price'}, inplace=True)
    
    df_clean['date'] = pd.to_datetime(df_clean['date'])
    df_clean['open_price'] = df_clean['open_price'].astype(float)
    
    df_clean = df_clean.sort_values('date').reset_index(drop=True)
    
    df_clean.to_csv(f'{symbol}_clean.csv', index=False)
    print(f"Saved {symbol}_clean.csv")

print("All files cleaned!")
```
