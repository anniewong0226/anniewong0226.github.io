```python
import requests
import json
import time
```

```python
API_KEY = '50PYVHBMHNN8UD6O'
BASE_URL = 'https://www.alphavantage.co/query'

# List of 6 stocks to download
symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA', 'NVDA']

# Loop through and download each
for symbol in symbols:
    print(f"Downloading {symbol}...")

    params = {
        'function': 'TIME_SERIES_DAILY',
        'symbol': symbol,
        'apikey': API_KEY,
        'outputsize': 'compact'  # Last 100 days
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    # Save to JSON file
    filename = f'{symbol}_data.json'
    with open(filename, 'w') as f:
        json.dump(data, f, indent=2)

    print(f"Saved {filename}")

    # IMPORTANT: Wait 12 seconds between requests (free tier limit is 5 calls/min)
    if symbol != symbols[-1]:  # Don't wait after last one
        print("Waiting 12 seconds...")
        time.sleep(12)

print("All downloads complete!")
```
